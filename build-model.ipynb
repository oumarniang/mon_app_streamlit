```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Charger les données
df = pd.read_csv("automobile_data.csv", sep=";")
df 
```

```python
df.dtypes
```

```python
# Sélectionner uniquement les colonnes numériques
df_num = df.select_dtypes(exclude="object")

# Vérifier si "price" est bien dans df_num, sinon l'extraire de df
if "price" in df_num.columns:
    y = df_num["price"]
    X = df_num.drop(columns=["price"])
else:
    y = df["price"]
    X = df_num

# Diviser les données en train et test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
df.dtypes
```

```python
df.describe().round(2)
```

```python
# Modèle de régression linéaire multiple

lr = LinearRegression()
lr.fit(X_train, y_train)

# Prédiction
y_pred = lr.predict(X_test)

# Évaluation du modèle
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Définition de la graine (seed)
seed = 123  

# Modèle de forêt aléatoire
rf = RandomForestRegressor(random_state=seed)
rf.fit(X_train, y_train)

# Prédiction
y_pred_rf = rf.predict(X_test)

# Évaluation du modèle
mse_rf = mean_squared_error(y_test, y_pred_rf)
print("Mean Squared Error (Random Forest):", mse_rf)
```

```python
mse_rf < mse # On retient le modèle rf
```
